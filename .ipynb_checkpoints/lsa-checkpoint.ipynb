```python
import json
import math
import numpy as np
from sympy import * 
from datetime import *
tk=10
query_list=[]
doc2id={}
word2id={}
file=open("query_list.txt","r").readlines()
for item in file:
    query_list.append(item.strip("\n"))
doc_list=[]
file=open("doc_list.txt","r").readlines()
doclen=0
for item in file:
    doc_list.append(item.strip("\n"))
    doc2id.update({item.strip("\n"):doclen})
    doclen+=1

i=0
query_w={}
for query in query_list:
    query_w.update({query:[]})
    file=open("Query/"+query,"r").readlines()
    for item in file:
        item=item.split()[:-1]
        for word in item:
            query_w[query].append(word)
            if word not in word2id.keys():
                word2id.update({word:i})
                i+=1
print(query_w)
for doc in doc_list:
    file=open("Document/"+doc,"r").readlines()[3:]
    for item in file:
        item=item.split()[:-1]
        for word in item:
            if word not in word2id.keys():
                word2id.update({word:i})
                i+=1
print(word2id)
```

```python
w_tk=np.zeros((tk,i))
w_tk+=1/tk
tk_doc=np.zeros((doclen,tk))
tk_doc+=1/doclen
print(tk_doc)
tk_w_doc=np.zeros((tk,i,doclen))
print(tk_w_doc)
```

```python
count_w_doc=np.zeros((i,doclen))
for doc in doc2id.keys():
    file=open("Document/"+doc,"r").readlines()[3:]
    for item in file:
        item=item.split()[:-1]
        for word in item:
            count_w_doc[ word2id[word],doc2id[doc]  ]+=1
```

```python
run=1
while True:
    t= datetime.now()
    for tk_no in range(tk):
        for w_no in range(i):
            for doc_no in range(doclen):
                tk_w_doc[tk_no,w_no,doc_no]=( w_tk[tk_no,w_no]*tk_doc[doc_no,tk_no]  )/np.sum( w_tk[:,w_no]*tk_doc[doc_no,:] )

    for tk_no in range(tk):
        for w_no in range(i):
            for doc_no in range(doclen):
                w_tk[w_no,tk_no]=np.sum(count_w_doc[w_no,:]*tk_w_doc[tk_no,w_no,:]  )/np.sum(count_w_doc*tk_w_doc[tk_no,:,:])

    for tk_no in range(tk):
        for w_no in range(i):
            for doc_no in range(doclen):
                tk_doc[doc_no,tk_no]=np.sum( count_w_doc[:,doc_no]*tk_w_doc[tk_no,:,doc_no]  )/doclen
    run+=1
    print(datetime.now()-t)
    if run>10:
        break
```

```python
bglm={}
file=open("BGLM.txt","r")
for item in file:
    item=item.split(" "+" "+" ")
    bglm.update({item[0]:item[1].strip("\n")})

print(bglm)
```

```python
alpha=0.3
beta=0.4
```

```python
query_ans0={}
query_ans={}
f=open("submission.txt","w")
f.write("Query,RetrievedDocuments\n")
for query in query_list:
    f.write(query+",")
    for doc in doc2id.keys():
        val=0
        for word in query_w[query]:
            val*=alpha*w_doc[doc2id[doc],word2id[word]]+beta* ( w_tk[:,word2id[word]]*tk_doc[doc2id[doc],:] ).sum()+(1-alpha-beta)*bglm[word]
        query_ans.update{doc:val}
    final=sorted( query_ans,key=lambda x:x[1],reverse=True    )

    for item in final:
        f.write(item+" ")
    f.write("\n")
    query_ans=query_ans0
f.close()
```
